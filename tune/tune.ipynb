## RNN MAPE 诊断（GPU，~2–3 min）

对比 **teacher-forcing val**（当前 tune/work 指标）与 **rolling val/test**（与竞赛提交一致）。

| 指标 | 含义 |
|------|------|
| `mape_val_teacher` | 验证窗内 y 用真值（偏乐观） |
| `mape_val_roll` | 验证段滚动预测（公平） |
| `mape_test_roll` | 测试段滚动预测 |

若 teacher ≈ 1.8% 而 roll ≈ 6%，说明 gap 来自评估协议而非模型训坏。

````python
import sys
from pathlib import Path
COMP = Path('..').resolve()
sys.path[:0] = [str(COMP), str(Path.cwd().resolve())]

from lib_clean import load, PATH_TRAIN, PATH_TEST
from lib_tune import setup_torch, prep_rnn_data, diagnose_rnn_torch
from lib_comp import load_params
from lib_feat import TIME, TARGET

params = load_params()
DEVICE = setup_torch()
df_hist, df_test = load(PATH_TRAIN, PATH_TEST)
df_all = __import__("pandas").concat([df_hist, df_test]).sort_values(TIME).reset_index(drop=True)
data = prep_rnn_data(df_all, params["val_start"], params.get("rnn_profile", "v1"))
diag_df, val_df, test_df = diagnose_rnn_torch(data, df_test, params["rnn"], device=DEVICE)
````

````python
import matplotlib.pyplot as plt
import numpy as np

kind = 'gru'
val_zoom = 288  # 3 days @ 15min
test_zoom = 672  # 7 days

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

vd = val_df.tail(val_zoom)
ax = axes[0]
ax.plot(vd[TIME], vd[f'{kind}_teacher_val'], label='teacher', alpha=0.8)
ax.plot(vd[TIME], vd[f'{kind}_roll_val'], label='roll', alpha=0.8)
ax.set_title(f'val last {val_zoom//96}d ({kind})')
ax.legend(fontsize=8)
ax.grid(True, alpha=0.3)

td = test_df.tail(test_zoom)
ax = axes[1]
if TARGET in td.columns:
    ax.plot(td[TIME], td[TARGET], 'k-', lw=2, label='true', alpha=0.85)
ax.plot(td[TIME], td[f'{kind}_roll_test'], label='roll test', alpha=0.8)
ax.set_title(f'test last {test_zoom//96}d ({kind})')
ax.legend(fontsize=8)
ax.grid(True, alpha=0.3)

ax = axes[2]
metrics = ['mape_val_teacher', 'mape_val_roll', 'mape_test_roll']
x = np.arange(len(metrics))
w = 0.35
for i, k in enumerate(diag_df['kind']):
    vals = [diag_df.loc[diag_df['kind']==k, m].iloc[0] for m in metrics]
    ax.bar(x + i * w, vals, width=w, label=k)
ax.set_xticks(x + w / 2)
ax.set_xticklabels(['val\nteacher', 'val\nroll', 'test\nroll'], fontsize=8)
ax.set_ylabel('MAPE %')
ax.set_title('MAPE comparison')
ax.legend(fontsize=8)
ax.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
````
